```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# How to scrape tabular data using beautifulsoup.

url= 'https://finance.yahoo.com/crypto/'
r = requests.get(url)
soup = BeautifulSoup(r.text,'lxml')
```

```python
table = soup.find('table',class_ = 'W(100%)')

headers = table.find_all('th')
titles = []
for i in headers:
    header = i.text
    titles.append(header)
titles
```

```python
df = pd.DataFrame(columns=titles)
df
```

```python
rows = table.find_all('tr')
for i in rows[1:]:
#     first_td = i.find_all('td')[0].find('div',class_= '___').text.strip() ==> we you faced a spaces in the first cell in the row.
    data = i.find_all('td') # and you will add here [1:]
    row = [tr.text for tr in data]
#     row.insert(0,first_td)
    l = len(df)
    df.loc[l] = row
```
