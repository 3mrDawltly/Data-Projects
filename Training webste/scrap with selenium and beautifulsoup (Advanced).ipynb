```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def scroll_to_element(driver, element):
    # Scroll until the element is found or a certain number of attempts are made
    max_attempts = 10
    attempts = 0

    while attempts < max_attempts:
        try:
            # Check if the element is visible
            if element.is_displayed():
                break
        except Exception as e:
            pass

        # Scroll down using JavaScript
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        # Increment the attempt counter
        attempts += 1

    if attempts == max_attempts:
        print("Element not found after scrolling.")
```

```python
url = 'https://webscraper.io/test-sites/e-commerce/ajax/computers/laptops'
driver = webdriver.Chrome()
driver.get(url)
```

```python
name_list = []
price_list = []
description_list = []
reviews_counts_list = []
rate_list = []

while True:
    html_source = driver.execute_script("return document.documentElement.outerHTML")
    soup = BeautifulSoup(html_source, 'html.parser')
    boxes = soup.find_all('div', class_='card-body')
    
    for j in range(len(boxes)):
        box = boxes[j]
        
        product_link = box.find('a', class_='title').get('href')
        product_full_link = 'https://webscraper.io' + product_link
#         print(product_full_link)
        
        # Fetch the content of the product page using requests
        product_page_content = requests.get(product_full_link).content
        product_page = BeautifulSoup(product_page_content, 'html.parser')

        name = product_page.find('h4', class_='title')
        name_list.append(name.text if name else None)

        price = product_page.find('h4', class_='price')
        price_list.append(price.text if price else None)

        description = product_page.find('p', class_='description')
        description_list.append(description.text if description else None)

        review_counts = product_page.find('p', class_='review-count')
        reviews_counts_list.append(review_counts.text.strip())

        rates = len(product_page.find_all('span', class_='ws-icon ws-icon-star'))
        rate_list.append(rates)
        

    try:
        # Find the element you want to scroll to (replace this with your own locator)
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'body > div.wrapper > div.container.test-site > div > div.col-lg-9 > div.pager > button.btn.btn-default.next'))
        )
        # Scroll to the element
        scroll_to_element(driver, next_button)
        # Now, you can interact with the found element or perform further actions
        next_button.click()

    except Exception as e:
        print(f"done")
        break
```

```python
df = pd.DataFrame({'Name':name_list,'Price':price_list,'Descriptions':description_list,'Reviews':reviews_counts_list,'Rates':rate_list})
df
```
